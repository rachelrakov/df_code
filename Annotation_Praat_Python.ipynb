## Separate out some useful things

```python
wordlist = []
silences = []
for i in range(length):
    word = words[i].text
    if word == "#":
        silences.append(word)
    else:
        wordlist.append(word)
print(wordlist)
print ("\n")
print (len(silences))
```

## What if you want to link your times and words as an indexable entry?

```python
txtgrd_info = []
for i in range(length):
    start = words[i].start_time
    end = words[i].end_time
    word = words[i].text
    entry = (start, end, word)
    txtgrd_info.append(entry)

print (txtgrd_info)
```

## What if you want to calculate how much silence you had in your file?

```python
total_silence = []
for i in txtgrd_info:
    if i[2]=="#":
        silence_time = i[1]-i[0]
        total_silence.append(silence_time)
print (total_silence)
print ("\n")
total_silence = sum(total_silence)

print (total_silence)
```
